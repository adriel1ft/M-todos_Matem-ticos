```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import limit_seq, symbols, S, oo
from sympy import *
from prettytable import PrettyTable
```

```python
n, x, y, z = symbols('n x y z')
```

```python
# Termo geral que vamos utilizar para os exemplos.
a_n = (n-1)/(n+1)
```

```python
# (a) o usuário não sabe se a sequência é convergente para um limite
```

```python
# Função que verifica se é convergente ou divergente
def verify_convergente(termo_geral):
  limit = limit_seq(termo_geral,n)

  if limit == S.Infinity or not(limit):
    return False
  else:
    return True
```

```python
if (verify_convergente(a_n)):
  print(f'Convergente. Limite L = {limit_seq(a_n)}')
else:
  print('Divergente')
```

```python
def opcao_a(termo_geral, nmin, nmax):
  n_values = list(range(nmin, nmax+1))
  a_values = [termo_geral.subs('n',i).evalf() for i in n_values] #substitui cada valor de n por i

  table = PrettyTable(['n','a(n)'])

  for n, i in zip(n_values, a_values):
    table.add_row([n, i])

  print(table)

  plt.scatter(n_values, a_values, marker ='o')
  plt.xlabel('n')
  plt.ylabel('a(n)')
```

```python
# saída da opção a)
opcao_a(a_n, 11, 30)
```

```python
# (b) o usuario sabe que a sequeência é covergente para um limite L
# já verificamos que L = 1 e manuscritamente achamos que n >= N(ε) <=> n >= 1999
```

```python
def opcao_b(termo_geral, nmin, nmax, L):
  n_values = list(range(nmin, nmax+1))
  a_values = [termo_geral.subs('n',i).evalf() for i in n_values]

  table = PrettyTable(['n','a(n)'])

  for n, i in zip(n_values, a_values):
    table.add_row([n, i])

  print(table)

  # segmentos de retas
  y_line1 = L + 1e-3
  y_line2 = L
  y_line3 = L - 1e-3


  plt.scatter(n_values, a_values, color='b', marker='o')

  plt.axhline(y=y_line1, linestyle='--', color='g', label='y = L + ε')
  plt.axhline(y=y_line2, linestyle='--', color='black', label="y' = L")
  plt.axhline(y=y_line3, linestyle='--', color='r', label="y'' = L - ε")

  plt.xlabel('n')
  plt.ylabel('a(n)')

  plt.legend()

```

```python
opcao_b(a_n,1999,1999+10,1)
```
